Load the Clean File

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("../data/processed/Clean_Superstore.xlsx")
```

Quick Structure Check

```python
df.shape #shows a tuple with the number of rows and columns
df.columns #shows the names of the columns
df.info() #provides a concise summary of the DataFrame, including the number of non-null entries and data types, 
# as well as memory usage, and the number of rows and columns.
df.describe() #provides descriptive statistics for numerical columns, including count, mean, standard deviation, min, max, and quartiles.
```

Find Problems

```python
df.isnull() #returns a DataFrame of the same shape as the original, with True for missing values and False for non-missing values
df[df['Profit']<0].head() #returns the first few rows of the DataFrame where the 'Profit' column is less than 0
df[df['Discount']>0.5] #returns the first few rows of the DataFrame where the 'Discount' column is greater than 1.0
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date']) #converts the 'Order Date' column to datetime format
df = df.dropna() #drops rows with any missing values
```

```python
df.to_csv("../data/processed/Clean_Superstore.csv", index=False)
# Saves the DataFrame to a CSV file without the index column.
df.to_excel("../data/processed/Clean_Superstore.xlsx", index=False)
# Saves the DataFrame to an Excel file without the index column.
```

EDA 

```python
df.columns.to_list()
```

```python
Region_Segment_Profit = df.groupby(['Segment', 'Region '])['Profit'].sum().reset_index()
# Groups the DataFrame by 'Region' and 'Segment', sums the 'Profit' for each group, and resets the index.
print(Region_Segment_Profit.sort_values(by='Profit'))
# Sorts the resulting DataFrame by 'Profit' in ascending order and prints it.
```

```python
discount_profit = df.groupby('Discount')[['Profit', 'Sales']].mean().reset_index()
# Groups the DataFrame by 'Discount', calculates the mean of 'Profit' and 'Sales' for each discount level, and resets the index.
print(discount_profit.sort_values(by='Discount'))
# Sorts the resulting DataFrame by 'Discount' in ascending order and prints it.
```

```python
plt.plot(discount_profit['Discount'], discount_profit['Profit'])# Plots 'Profit' against 'Discount'.
plt.title("Average profit by Discount")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.grid(True)# Adds a grid to the plot for better readability.
plt.show()# Displays the plot.
#this shows, as the discount increases, the profit decreases, which is expected as higher discounts lead to lower profits.
```

To check the most Unprofitable Product

```python
loses = df[df['Profit']<0] # Filters the DataFrame to include only rows where 'Profit' is less than 0.
loss_summary = loses.groupby('Sub-Category')['Profit'].sum().sort_values()
print(loss_summary)
# Groups the filtered DataFrame by 'SubCategory', sums the 'Profit' for each subcategory, sorts the results, and prints it.
#This shows, the most loses are for Blinders, Tables, Machiens and Bookcases, which are the most expensive items in the store.
```

```python
df = df[df['Region '] != 'CATEGORY CLEAN']
# Filters the DataFrame to exclude rows where 'Region' is 'CATEGORY CLEAN'.
```

```python
region_profit = df.groupby('Region ')['Profit'].sum().sort_values()

# Groups the DataFrame by 'Region', sums the 'Profit' for each region, sorts the results, and prints it.
region_profit.plot(kind='bar', title="Profit by Region", xlabel="Region", ylabel="Profit")
plt.grid(True)
plt.show()


# Plots a bar chart of 'Profit' by 'Region', sets the title and labels for the axes, adds a grid, and displays the plot.
#This shows, the most profit is made in the West region, followed by the East and Central regions.

```
